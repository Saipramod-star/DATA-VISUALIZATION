```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,15,5)
y = 2*x+5
y1 = 5*x+5
fig = plt.figure(figsize=(5,5))
plt.plot(x,y,linewidth = 3,linestyle = ":",color = "red",alpha = 1,marker = "o")
plt.plot(x,y1,linewidth = 3,linestyle = ":",color = "black",alpha = 2,marker = "x")
plt.title("MY FIRST LINE PLOT")
plt.xlabel("X-AXIS",color = "blue")
plt.ylabel("Y-AXIS")
plt.legend(["line x","line y"],loc = "best")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = {"apples":20,"banana":15,"orange":23,"cucumber":6,"grapes":17}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (5,5))
plt.barh(names,values,color= "yellow",alpha = 1)
plt.title("my first bar")
plt.xlabel("FRUITS",color = "black")
plt.ylabel("QUANTITY",color = "blue")
plt.savefig("GRAPHS.PNG")
plt.show()
```

SCATTER PLOT

```python
import matplotlib.pyplot as plt
%matplotlib inline

a = [10,20,50,40,20,50]
b = [1,2,3,2,5,4]
x = [3,1,5,6,7,3]

plt.figure(figsize=(3,3))
plt.scatter(a,b)
plt.scatter(a,x)
plt.savefig("GRAPHS.PNG")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

a = [10,20,50,40,20,50]
b = [1,2,3,2,5,4]
x = [3,1,5,6,7,3]

plt.scatter(a,b,color = "black",s=100,edgecolors="black",marker="o",alpha=1)
plt.scatter(a,x,color = "blue",s=200,edgecolors="black",marker="o",alpha=1)
plt.legend(["b","x"],loc = "best")
plt.title("SCATTER PLOT")
plt.xlabel("A label")
plt.ylabel("X label")
plt.grid(True)
plt.savefig("GRAPHS.PNG")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

total = [10,20,1,30,40,20,20,60,70]
order = [10,2,4,3,6,10,21]
discount = [10,20,35,30,60,50,60]
data = list([total,order,discount])

plt.violinplot(data,showmeans = True,showmedians = True)
plt.title("BOX PLOT")
plt.grid(True)
plt.show()
```

PIE CHART

```python
import matplotlib.pyplot as plt
%matplotlib inline

label = ["dog","cat","tiger","lion"]
size = [25,30,50,60]
plt.pie(size,labels = label,autopct="%1.2f%%",shadow = False,startangle = 360,explode=(0,0.1,0,0))
plt.title("PIE CHART")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

label = ["dog","cat","tiger"]
size = [25,30,50]
size_centre = [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]

pie1 = plt.pie(size,labels = label,radius = 1.5,autopct="%1.2f%%",colors = [a(0.5),b(1.5),c(1.5)])
pie2 = plt.pie(size_centre,radius = 0.5,colors = "r")
plt.show()
```
